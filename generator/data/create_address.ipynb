```python
from geopy.geocoders import Nominatim
import pandas as pd
from tqdm import tqdm
geolocator = Nominatim(user_agent="adresse_fr")
villes = list(pd.read_csv('ville.csv')['0'])
```

```python
addresses = []
```

```python
villes = villes[:1000] + ['Paris']
for city in tqdm(villes):
    city += ', France'
    locations = geolocator.geocode(city, exactly_one=False, limit=100)
    if locations:
        for location in locations:
            latitude, longitude = location[-1][0], location[-1][1]
            addresses.append(geolocator.reverse((latitude, longitude), exactly_one=True)[0])

    
```

```python
addresses
```

```python
df = pd.DataFrame({'address' : addresses})
df.to_csv('address.csv', index=False)
```
